```python
import numpy as np                # Librairie pour le calcul scientifique
import matplotlib.pyplot as plt   # Librairie graphique
```

```python
a_file = open("eta.txt", "r") # Ouvrir le fichier

x = []
y = []
for line in a_file: # Lire chaque ligne
    abscisse, value = line.split(' ')
    x.append(float(abscisse))
    y.append(float(value))
a_file.close()

plt.figure(figsize=[15,9])
plt.plot(x,y,label="Champ au eta retenu")
plt.plot(x,np.zeros(len(x)),label="y = 0")
plt.title("Champ au point eta retenu pour chaque x discrétisé")
plt.legend()
plt.show()
```

```python

x = np.arange(-5,5,0.1) 
y = np.arange(-5,5,0.1) 
  
X,Y = np.meshgrid(x,y) 
  
Ex = (X + 1)/((X+1)*2 + Y**2) - (X - 1)/((X-1)*2 + Y*2) 
Ey = Y/((X+1)*2 + Y**2) - Y/((X-1)*2 + Y*2) 
  
plt.figure(figsize=(10, 10)) 
plt.streamplot(X,Y,Ex,Ey, density=1.4, linewidth=None, color='#A23BEC') 
plt.plot(-1,0,'-or') 
plt.plot(1,0,'-og') 
plt.title('Electromagnetic Field') 
  
plt.grid() 
plt.show()
```
